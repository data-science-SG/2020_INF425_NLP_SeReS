#Pegando vários tweets de um mesmo usuário

```
def tweets_from_user(user, depression, count=5):
  tweets =  api.user_timeline(id=user,
                      tweet_mode = 'extended',
                      include_rts = False,
                      exclude_replies=True,
                      contributor_details=False,
                      include_entities=False,
                      count=count)


  Tweets_User = pd.DataFrame()
  if len(tweets) > min_tweets:
    users_locs = [[tweet.user.screen_name, tweet.full_text] for tweet in tweets]

    Tweets_User =  pd.DataFrame(data=users_locs, 
                      columns=['usuario', 'texto'])
    
    Tweets_User['depressao'] = depression 

  return Tweets_User
```

#Traduzindo os tweets de portgues para inglês ;)

https://towardsdatascience.com/translate-a-pandas-data-frame-using-googletrans-library-fb0aa7fca592

```
# # install googletrans using pip
!pip install googletrans

# import the library
import googletrans
from googletrans import Translator
```

```
tradutor = Translator()
```

```
def traduz (doc):

  traduzido = tradutor.translate(doc, src='pt', dest='en').text

  return traduzido
```

```
df_final['texto_traduzido'] = df_final['texto'].apply(lambda x: traduz(x))
```

```
#Excluindo os que não foram traduzidos
df_final.query('texto == texto_traduzido')

a = df_final.query('texto == texto_traduzido').index
df_final.drop(a, inplace=True)
```
